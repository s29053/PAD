4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
import pandas as pd

orders = pd.read_csv("orders.csv")
orders.head()
```

```python
orders.describe
```

```python
orders.info
```

```python
# Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień

orders['order_date']= pd.to_datetime(orders['order_date'], format='%Y/%m/%d')
orders.head(5)
```

```python
# Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category

print("Unique values:")
print(orders['tshirt_category'].unique())
print("Number of unique values:")
print(orders['tshirt_category'].nunique())
```

```python
# Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że 
# dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko 
# jedna wersja white. Skorzystaj z klasy string i metody apply

orders['tshirt_category'] = orders['tshirt_category'].apply(lambda x: x.lower())
orders['tshirt_category'] = orders['tshirt_category'].str.replace(" shirt", " t-shirt")
orders['tshirt_category'] = orders['tshirt_category'].str.replace(" tshirt", " t-shirt")
orders['tshirt_category'] = orders['tshirt_category'].str.replace("tennis t-shirt", "t-shirt")
orders['tshirt_category'] = orders['tshirt_category'].str.replace("wh ", "white ")
orders['tshirt_category'] = orders['tshirt_category'].str.replace("bl ", "black ")

orders1 = orders.copy()

orders['tshirt_category']
```

```python
# Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła 
# o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie 
# informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o
#  kolorze. Skorzystaj z metody apply.

def my_split(x):
  split1 = x.split(' ')
  if split1[0] in ["white", "black"]:
    return x
  return f"None {x} None"

orders['tshirt_category'] = orders['tshirt_category'].apply(lambda x: my_split(x))
orders[['tshirt_colour', 'tshirt_type', 'tshirt_size']] = orders['tshirt_category'].str.split(' ', expand=True)
orders[['tshirt_colour', 'tshirt_type',   'tshirt_size']]
```

```python
# Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.
mask = (orders['order_date'] > '2014-12-31') & (orders['order_date'] < '2016-8-2')
orders.loc[mask]
```

```python
customers = pd.read_csv("customers.csv")

# Wypisz kolumny

customers.columns
```

```python
# Porównaj rozmiar ramki z rozmiarem ramki w zadaniu 1

print("Rozmiar ramki orders (wiersze, kolumny):\n", orders.shape,"\n","Rozmiar ramki customers (wiersze, kolumny):\n", customers.shape)
```

```python
# Zmień nazwę kolumny customerID na customer_id upewniając się, że dane zmiana 
# zostanie zapisana w pamięci dzięki jednemu z argumentów metody do zmiany nazwy kolumny

customers.rename(columns={"customerID": "customer_id"}, inplace=True)
customers.columns
```

```python
# Połącz ramkę orders z ramką customers po kolumnie customer_id

new_df = orders.merge(customers, on="customer_id", how="left")


# Wykorzystałem metodę left join, ponieważ zwróci nam ona wszystkie zamówienia 
# wraz z danymi klienta, jeśli takie istnieją. Jesteśmy skupieni na zamówieniach,
# więc nie chcemy tracić rekordów dla których nie ma danych klienta, co popiera 
# użycie tej metody. Sam brak danych o klientach również może nieść pewną 
# informację, więc tym bardziej użycie metody left join wydaje się być słuszne.
```
